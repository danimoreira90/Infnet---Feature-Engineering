**Diferença entre MinMaxScaler e StandardScaler:**

**MinMaxScaler:** Esta técnica de escalonamento ajusta os valores de forma que fiquem dentro de um intervalo definido, tipicamente entre 0 e 1 (ou -1 a 1, se houver valores negativos na característica original). É útil quando precisamos de valores normalizados sem desviar a distribuição dos dados. No entanto, é sensível a outliers, pois eles podem comprimir a maioria dos dados em um pequeno intervalo.

**StandardScaler:** Esta técnica remove a média e divide pela variância, resultando em uma distribuição com uma média 0 e desvio padrão 1. Isso significa que, após o escalonamento, os valores da característica serão distribuídos em torno de 0, com cada unidade de desvio padrão correspondendo a uma faixa de valores na característica original. É menos sensível a outliers em comparação com o MinMaxScaler, pois não está confinando os valores a um intervalo específico.









#**Questão 5:**

A norma-l2, também conhecida como norma euclidiana ou magnitude de um vetor, é uma medida de distância em um espaço euclidiano. Ela representa a distância "direta" (como se fosse uma linha reta) entre o ponto de origem do vetor e seu ponto final. No contexto de uma característica de um dataset, calcular a norma-l2 de uma característica pode ajudar a entender sua magnitude geral em relação à origem.

A fórmula para calcular a norma-l2 de um vetor \( \mathbf{x} \) com componentes \( x_1, x_2, \ldots, x_n \) é:

\[
\| \mathbf{x} \|_2 = \sqrt{x_1^2 + x_2^2 + \ldots + x_n^2}
\]

Se estivermos tratando de uma única característica em um dataset com várias instâncias, você pode considerar cada valor da característica como um componente de um vetor. Assim, a norma-l2 dessa característica seria calculada tomando a raiz quadrada da soma dos quadrados de todos os seus valores.

**Para que serve?**

- **Normalização:** A norma-l2 é frequentemente usada para normalizar vetores. No contexto de machine learning e processamento de sinal, a normalização pode ser importante para comparar vetores de maneira justa, especialmente quando eles são de diferentes magnitudes. A normalização pela norma-l2 faz com que o vetor tenha um comprimento (ou magnitude) de 1, preservando a direção mas normalizando seu tamanho.
  
- **Regularização:** Em machine learning, particularmente em regressão linear e regressão logística, a norma-l2 é usada como base para a regularização L2 (Ridge). A regularização L2 penaliza os coeficientes do modelo proporcionalmente ao quadrado de sua magnitude, ajudando a prevenir o overfitting e a manter os pesos do modelo relativamente pequenos, promovendo modelos mais simples.

- **Medição de Distância:** A norma-l2 pode ser usada para medir a distância entre pontos em espaços de alta dimensão, o que é útil em várias aplicações, como na classificação de k-nearest neighbors (KNN) e em algoritmos de clustering como o k-means.

A aplicação da norma-l2 ajuda a garantir que os modelos sejam menos sensíveis a escala das características, além de contribuir para a estabilidade numérica de algoritmos de otimização.

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import make_scorer, accuracy_score, f1_score
```

```python
# Prepare the stopwords list and the lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_text(text):
    # Tokenize by splitting the text into words
    words = text.split()
    # Lemmatize each word
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(lemmatized_words)
```

```python
# Lemmatize the reviews in the DataFrame
dfimdb['lemmatized_review'] = dfimdb['review'].apply(lemmatize_text)
```

```python

# Define a TfidfVectorizer with Bag-of-nGrams
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
```

```python
# Create a pipeline with TfidfVectorizer and LogisticRegression
model = make_pipeline(vectorizer, LogisticRegression())
```

```python
# Define the features and labels
X = dfimdb['lemmatized_review']
y = dfimdb['sentiment']
```

```python
# Calculate accuracy and F1 score using cross-validation
accuracy = cross_val_score(model, X, y, scoring=make_scorer(accuracy_score), cv=5).mean()
f1 = cross_val_score(model, X, y, scoring=make_scorer(f1_score, average='weighted'), cv=5).mean()
```

```python
print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score: {f1:.4f}")
```

#**Questão 11:**

A Análise de Componentes Principais (PCA) é uma técnica de redução de dimensionalidade amplamente utilizada em estatísticas e aprendizado de máquina. O objetivo do PCA é transformar um conjunto de variáveis possivelmente correlacionadas em um conjunto menor de variáveis lineares não correlacionadas chamadas Componentes Principais (PCs). Essa transformação é feita de maneira que os primeiros PCs retenham a maior parte da variância presente no conjunto de dados original. Vamos entender o funcionamento do PCA e o papel dos autovalores e autovetores nesse processo.

### Funcionamento do PCA

1. **Normalização dos Dados:** PCA começa com a normalização dos dados. Isso é feito para garantir que cada característica contribua igualmente para o resultado, independentemente da sua escala.

2. **Cálculo da Matriz de Covariância:** Calcula-se a matriz de covariância dos dados normalizados para entender como as variáveis do conjunto de dados estão variando juntas.

3. **Cálculo de Autovalores e Autovetores:** A matriz de covariância é então decomposta para encontrar seus autovalores e autovetores. Esses são fundamentais para entender a direção e a magnitude da variância nos dados.

    - **Autovalores:** Representam a magnitude da variância capturada por cada componente principal. Um autovalor maior indica que mais variância é capturada por seu componente principal associado.
    
    - **Autovetores:** Indicam a direção no espaço original dos dados em que cada componente principal se estende. Em outras palavras, cada autovetor representa a direção de um PC no espaço de características original.

4. **Seleção dos Componentes Principais:** Os autovalores são usados para ordenar os componentes principais em termos de sua variância explicada. Os componentes principais são selecionados com base na quantidade de variância total que se deseja reter no conjunto de dados reduzido.

5. **Transformação dos Dados:** Finalmente, os dados originais são projetados nos autovetores selecionados (os PCs), transformando-os do espaço de características original para o novo espaço de características dos PCs.

### Autovalores e Autovetores nos PCs

Em PCA, os autovalores e autovetores derivam da matriz de covariância dos dados e desempenham papéis cruciais:

- **Autovalores:** Quantificam a variância (ou "importância") de cada componente principal. Um autovalor alto significa que o componente principal associado captura uma grande quantidade da variância total nos dados.

- **Autovetores:** Definem as direções dos componentes principais. Cada autovetor aponta na direção de máxima variância no espaço multidimensional dos dados. Quando os dados são projetados nesses autovetores (transformação para os PCs), resulta em um novo conjunto de coordenadas que são linearmente independentes (não correlacionadas).

Os PCs são ordenados por sua variância explicada (autovalores), com o primeiro PC capturando a maior variância, o segundo capturando a segunda maior variância, e assim por diante. Escolher os primeiros PCs permite reduzir a dimensionalidade dos dados enquanto se retém a maior parte da informação original.

O PCA, portanto, facilita a compressão de dados, visualização em dimensões mais baixas, e pode melhorar o desempenho de algoritmos de aprendizado de máquina ao remover ruídos ou informações redundantes.
